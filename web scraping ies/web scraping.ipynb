```python
# multiple products store into array of dictionaries
products_data = []
for url in store_product_urls:
    
    page = requests.get(url, headers=headers)
    soup = BeautifulSoup(page.content, "html.parser")

    desc_div = soup.find('div', itemprop="description")
    desc_tags = desc_div.find_all('p') if desc_div else []

    title = soup.find('h2', itemprop="name").get_text() if soup.find('h2', itemprop="name") else ""
    OEM = soup.find(itemprop="brand").get_text() if soup.find(itemprop="brand") else ""
    OEM_part_number = soup.find(itemprop="productID").get_text() if soup.find(itemprop="productID") else ""
    desc = desc_tags[0].get_text() if desc_tags else "No Description"
    #stock = desc_tags[1].get_text() if len(desc_tags) > 1 else "Not In Stock"
    stock = soup.find('div', class_="in-stock-message").get_text()[1:-1] if soup.find('div', class_="in-stock-message") else "Not In Stock"
    product_data = {
        'title': title,
        'OEM': OEM,
        'OEM Number': OEM_part_number,
        'Description': desc,
        'Product Availability': stock
    }
    products_data.append(product_data)

print(products_data)
```

```python
# import to csv
import csv

csv_file = 'more_semiconductor_csv.csv'

# Write the data to a CSV file
with open(csv_file, mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=products_data[0].keys())
    writer.writeheader()
    writer.writerows(products_data)

print(f'Data saved to {csv_file}')
```

```python
import sqlite3
import pandas as pd


df = pd.read_csv('semiconductor_csv.csv')

conn = sqlite3.connect('products.db')
cursor = conn.cursor()

cursor.execute('''
    CREATE TABLE IF NOT EXISTS products (
        title TEXT,
        OEM TEXT,
        OEM_part_number TEXT,
        description TEXT,
        stock_status TEXT
    )
''')

df.to_sql('products', conn, if_exists='replace', index=False)

conn.commit()
conn.close()

print("Data successfully written to products.db")
```
